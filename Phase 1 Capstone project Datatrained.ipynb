# Scraping  data using the given URL.

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException 

url = 'https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'

# Initialize the Selenium webdriver 
driver = webdriver.Chrome()

driver.get(url)

# Function to extract metascore from movie element
def extract_metascore(movie_element):
    try:
        metascore = movie_element.find_element(By.CSS_SELECTOR, '.metascore').text
    except NoSuchElementException:
        metascore = 'N/A'
    return metascore

# Wait for the page to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.lister-list')))

# Create empty lists to store the data
movie_names = []
director_names = []
durations = []
years = []
ratings = []
metascores = []

# Loop through all pages
while True:
    # Find all movie elements using XPATH
    movie_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]')

    # Loop through the movie elements and extract the data
    for movie_element in movie_elements:
        # Extract movie name
        movie_name = movie_element.find_element(By.XPATH, './/h3/a').text
        movie_names.append(movie_name)

        # Extract director name
        try:
            director_name = movie_element.find_element(By.CSS_SELECTOR, 'p:not(.text-muted):not(.text-small) a').text
        except NoSuchElementException:
            director_name = 'N/A'
        director_names.append(director_name)

        # Extract duration
        duration = movie_element.find_element(By.XPATH, './/p/span[@class="runtime"]').text
        durations.append(duration)

        # Extract year
        year = movie_element.find_element(By.XPATH, './/h3/span[2]').text.strip('()')
        years.append(year)

        # Extract rating
        rating = movie_element.find_element(By.XPATH, './/strong').text
        ratings.append(rating)

        # Extract Metascore
        metascore = extract_metascore(movie_element)
        metascores.append(metascore)

    try:
        # Click the "Next" button
        next_button = driver.find_element(By.XPATH, '//a[@class="lister-page-next next-page"]')
        next_button.click()
        wait.until(EC.staleness_of(movie_elements[0]))  # Wait for the page to load
    except NoSuchElementException:
        # If the "Next" button is not found, we have reached the last page, exit the loop
        break

# Create a DataFrame to store the data
data = {
    'Movie Name': movie_names,
    'Director Name': director_names,
    'Duration': durations,
    'Year': years,
    'Rating': ratings,
    'Metascore': metascores
}

df= pd.DataFrame(data)

# Print the DataFrame
print(df)

# Close the browser after scraping
driver.quit()
```

# Scraping  data set 2 using the given URL.

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException


url = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&ref_=adv_prv'

# Initialize the Selenium webdriver 
driver = webdriver.Chrome()
driver.get(url)

# Function to scroll the page to load all elements
def scroll_to_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    wait.until(EC.invisibility_of_element_located((By.CLASS_NAME, 'scroller-infinite')))
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.lister-item-content')))

# Wait for the page to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.lister-list')))
scroll_to_bottom(driver)

# Create empty lists to store the data
movie_names = []
stars = []
votes = []
genres = []
gross_collections = []
certifications = []

# Loop through all pages
while True:
    # Find all movie elements using XPATH
    movie_elements = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]')

    # Loop through the movie elements and extract the data
    for movie_element in movie_elements:
        # Extract movie name
        movie_name = movie_element.find_element(By.XPATH, './/h3/a').text
        movie_names.append(movie_name)

        # Extract stars
        try:
            star_elements = movie_element.find_elements(By.XPATH, './/p[contains(., "Stars:")]/a')
            star_info = [star_element.text for star_element in star_elements]
        except NoSuchElementException:
            star_info = ['N/A']
        stars.append(star_info)

        # Extract votes
        try:
            vote_element = movie_element.find_element(By.XPATH, './/p/span[@name="nv"]')
            vote = vote_element.get_attribute('data-value')
        except NoSuchElementException:
            vote = 'N/A'
        votes.append(vote)

        # Extract genres
        try:
            genre_element = movie_element.find_element(By.XPATH, './/p/span[@class="genre"]')
            genre_info = genre_element.text.split(', ')
        except NoSuchElementException:
            genre_info = ['N/A']
        genres.append(genre_info)

        # Extract gross collection
        try:
            gross_element = movie_element.find_element(By.XPATH, './/p[4]/span[5]')
            gross_collection = gross_element.text
        except NoSuchElementException:
            gross_collection = 'N/A'
        gross_collections.append(gross_collection)

        # Extract certification
        try:
            certification_element = movie_element.find_element(By.XPATH, './/span[@class="certificate"]')
            certification = certification_element.text
        except NoSuchElementException:
            certification = 'N/A'
        certifications.append(certification)

    try:
        # Click the "Next" button
        next_button = driver.find_element(By.XPATH, '//a[@class="lister-page-next next-page"]')
        next_button.click()
        scroll_to_bottom(driver)
    except NoSuchElementException:
        # If the "Next" button is not found, we have reached the last page, exit the loop
        break

# Create a DataFrame to store the data
data = {
    'Movie Name': movie_names,
    'Votes': votes,
    'Gross Collection': gross_collections,
    'Certification': certifications
}

# Create separate columns for stars (star1, star2, star3, star4, star5)
for i in range(1, 6):
    data[f'Star{i}'] = [star[i - 1] if i <= len(star) else 'N/A' for star in stars]

# Create separate columns for genres (genre1, genre2, genre3)
for i in range(1, 4):
    data[f'Genre{i}'] = [genre[i - 1] if i <= len(genre) else 'N/A' for genre in genres]

df1 = pd.DataFrame(data)

# Drop 'Star1' column
df1.drop(columns=['Star1'], inplace=True)

# Rename the remaining columns
df1.rename(columns={'Star2': 'Star1', 'Star3': 'Star2', 'Star4': 'Star3', 'Star5': 'Star4'}, inplace=True)

# Print the DataFrame
print(df1)

# Close the browser after scraping
driver.quit()
```
